```python
def transformCoopSourceFile(df,ProgramCodesFileDF,coopTargetFileDF):
    df.columns=df.columns.str.strip()
    ProgramCodesFileDF.columns=ProgramCodesFileDF.columns.str.strip()
    coopTargetFileDF.columns = coopTargetFileDF.columns.str.strip()
    
    #df=df.rename(columns={'Name':'Old_Name'})
    column_mapping = {
    "NUID": "Khor/id"

}
    transformeddf=df.rename(columns=column_mapping)
    transformeddf["Key company"]=''
    transformeddf["Count groups"]=0
    transformeddf['Khor/id']=transformeddf['Khor/id'].astype(str).str.zfill(9)
    
    
    #transformeddf["First Name, Last Name :"]=transformeddf["First Name:"]+' '+transformeddf["Last Name:"]
    #transformeddf["Race and/or Ethnicity"]=''
    #transformeddf["Reason buckets"]=df["Asking for Help with : (Primary Reason)"]
    #transformeddf["Drop-in  or appointmnet"]='Appointment'
    #transformeddf['NUID#:']=transformeddf['NUID#:'].astype(str).str.zfill(9)
    
    #Drop un-necessary columns
    transformeddf.drop(columns=["Degree Level","Global Agency","ID","Symplicity Job ID","New Job Title2","Campus","CampusD","Primary Co-op Advisor","Job Manager","School","Primary College","New Job Title","Job","Archived","Archived On"],axis=1)
    
    
    transformeddf=transformeddf.merge(ProgramCodesFileDF,left_on="Program Code", right_on='Program Code', how='left')
    #transformeddf["Select a program that you are currently pursuing :"]=transformeddf["Select a program that you are currently pursuing :"]
    
    #transform target ID
    coopTargetFileDF['Khor/id']=coopTargetFileDF['Khor/id'].astype(str).str.zfill(9)
    
    columns_to_compare = ['Khor/id', 'Work Term', 'Email','Employer']
    #transformeddf['Existing']=transformeddf.apply(lambda row:1 if row[columns_to_compare].values in careerAdvisingTargetFileDF[columns_to_compare].values else 0, axis=1)
    
    #***********
    
    #identify duplicates between Source and target files and flag them as Existing, 
    # Filter out all duplicate values and only return existing records.
    
    #****************
    
    # Filter Work Terms older than Fall-2023 from comparison
    
    transformeddf = transformeddf[~transformeddf['Work Term'].isin(["2022 - Summer", "2022 - Fall", "2023 - Summer", "2022 - Spring", "2023 - Spring", "2022 - Winter", "2023 - Winter"])]
    coopTargetFileDF = coopTargetFileDF[~coopTargetFileDF['Work Term'].isin(["2022 - Summer", "2022 - Fall", "2023 - Summer", "2022 - Spring", "2023 - Spring", "2022 - Winter", "2023 - Winter"])]
    
    # Concatenate the values in the specified columns for both dataframes
    transformeddf['Concatenated'] = transformeddf[columns_to_compare].astype(str).agg(''.join, axis=1)
    coopTargetFileDF['Concatenated'] = coopTargetFileDF[columns_to_compare].astype(str).agg(''.join, axis=1)
    # Add a new column 'Existing' to flag matching rows
    transformeddf['Existing'] = transformeddf['Concatenated'].isin(coopTargetFileDF['Concatenated']).astype(int)
    # Drop the 'Concatenated' column
    #transformeddf = transformeddf.drop(columns=['Concatenated'])
    #print(transformeddf['Existing'].value_counts())
    
    
    pd.set_option('display.max_colwidth', None)
    print(coopTargetFileDF[coopTargetFileDF['Khor/id']=='002631568']['Concatenated'])
    print(transformeddf[transformeddf['Khor/id']=='002631568']['Concatenated'])
    
    
    
    # Return only Non-Existing Values
    finaldf=transformeddf.loc[transformeddf['Existing']==0].copy()
    
    finaldf["Wage"] = pd.to_numeric(finaldf["Wage"], errors='coerce')
    
    # Fill hourly rate value for new records
    conditions = [
    (finaldf["Wage Type"] == "Hourly"),
    (finaldf["Wage Type"] == "Monthly"),
    (finaldf["Wage Type"] == "Unpaid"),
    (finaldf["Wage Type"] == "Weekly"),
    (finaldf["Wage Type"] == "Stipend"),
    (finaldf["Wage Type"] == "Unspecified")
     ]

    choices = [
    finaldf["Wage"],
    finaldf["Wage"] / 173,
    finaldf["Wage"],
    finaldf["Wage"] / 40,
    finaldf["Wage"] / 173,
    finaldf["Wage"]
     ]


    finaldf["Hourly rate"] = np.select(conditions, choices, default=None)
    
    #return merged_df

    return finaldf
```

```python
def loadCoopSourceFile(finaldf,coopTargetFileDF):
    coopTargetFileDF.columns = coopTargetFileDF.columns.str.strip()

    desired_column_order = [
    "College", "Khor/id", "Work Term", "First Name", "Last Name", "Email",
    "Student Primary Manager", "Major", "Graduation Date", "Citizenship Type",
    "Resident Status", "Visa Type", "Visa Status", "Work Authorization",
    "Parent Organization", "Employer", "Key company", "New Job Title",
    "Placement Length", "Start Date", "End Date", "Hours Per Week",
    "Wage Type", "Wage", "Hourly rate", "Employment Status",
    "Is this a Self-Developed or Global Agency position?", "City", "State",
    "Country", "Will you be working remotely for any portion of this placement?",
    "Remote City", "Remote State", "Remote Country", "Is your supervisor known at this time?",
    "Supervisor Email", "Supervisor Name", "Job Contact", "Job Contact Email",
    "Approved Date", "Location", "Gender", "Program Code", "Program",
    "Program full name", "Count groups", "Citizenship Country"
]

# Reorder the columns in the DataFrame
    finaldf = finaldf[desired_column_order]
    #finaldf['College']=finaldf['College'].replace({'Khoury Coll of Comp Sciences':'Khoury College','Coll of Professional Studies':'College of Professional Studies'})
    
    #targetdf=coopTargetFileDF.append(finaldf)
    #targetdf = targetdf.drop(columns=['Concatenated'])
    targetdf= pd.concat([coopTargetFileDF, finaldf], ignore_index=True)
    targetdf["Start Date"]=pd.to_datetime(targetdf["Start Date"]).dt.strftime('%m/%d/%Y')
    targetdf["End Date"]=pd.to_datetime(targetdf["End Date"]).dt.strftime('%m/%d/%Y')
    
    targetdf['Count groups'] = targetdf.groupby('Khor/id')['Khor/id'].transform('count')
    
    return targetdf
```
