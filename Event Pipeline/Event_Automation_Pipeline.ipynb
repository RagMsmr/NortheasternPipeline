```python
def fn_importfile(path, sheetName):
    for file in path:
        sourcefile = pd.ExcelFile(file)
        df=pd.read_excel(sourcefile,sheetName, dtype=str)
        return df
```

```python
sourceEventFileDF = fn_importfile(SourceEventsFilePath, SourceCarrerSheetName)
campusStudentsFileDF = fn_importfile(CampusStudentsFilePath, CampusStudentsSheetName)
MappingFileDF = fn_importfile(MappingConstantsFilePath, MappingSheetName)
ConstantsFileDF = fn_importfile(MappingConstantsFilePath, ConstantsSheetName)
#targetEventsFileDF = fn_importfile(TargetEventsFilePath, 'Sheet1')
```

```python
def trimLeadingTrailingSpaces(df):
    df.columns = df.columns.str.strip()
    df=df.apply(lambda x:x.str.strip() if x.dtype=='object' else x)
    return df
```

```python
def fn_createDict(df, keyName1, keyName2=""):
    nested_dict = {}
    for index, row in df.iterrows():
        if keyName2:
            key_value = row[keyName1] + " " + row[keyName2]
        else:
            key_value = row[keyName1]
            # Remove the key column from the row
            row = row.drop(keyName1)
        # Convert the row to a dictionary
        row_dict = dict(row)
        # Add the row dictionary to the nested dictionary with the key value as the key
        nested_dict[key_value] = row_dict
    return nested_dict

```

```python
#source_event_dict = fn_createDict(sourceEventFileDF, "Student ID#:")
students_studentId_dict = fn_createDict(campusStudentsFileDF, "Id")
students_name_dict = fn_createDict(campusStudentsFileDF, "First Name", "Last Name")
students_email_dict = fn_createDict(campusStudentsFileDF, "Student Email Address")
constants_dict = fn_createDict(ConstantsFileDF, 'TargetColumnName')
```

```python
def getValueFromStudentRegistrarFile(row, student_file_column_name):
  if(student_file_column_name == ''):
    print('Empty column name')
  student_id = str(row['Student ID#:']).zfill(9)
  student_id
  name = row['First Name'].strip() + " " + row['Last Name'].strip()
  #print(row['Northeastern Email address:'] + 'row= '+ str(student_file_column_name))
  email = row['Northeastern Email'].strip()
  if student_id:
    rowValue = students_studentId_dict.get(student_id)
    if rowValue:
        return rowValue.get(student_file_column_name)
  elif name:
      rowValue = students_name_dict.get(name)
      if rowValue:
          return rowValue.get(student_file_column_name)
  elif email:
      rowValue = students_email_dict.get(email)
      if rowValue:
          return rowValue.get(student_file_column_name)
  else:
    return None
```

```python
def generate_target_file(MappingFileDF, sourceEventFileDF):
    target_file = {}

    # Iterate through each row in sourceEventFileDF
    for index, row in sourceEventFileDF.iterrows():
        target_row = {}
        first_name = row.get('First Name') or row.get('First Name:') or getValueFromStudentRegistrarFile(row, 'First Name')
        last_name = row.get('Last Name') or row.get('Last Name:')or getValueFromStudentRegistrarFile(row, 'Last Name')
        target_row['Name (Original Name)'] = first_name + " " + last_name
        # Iterate through each row in MappingFileDF
        for mapping_index, mapping_row in MappingFileDF.iterrows():
            source_column_name = mapping_row['SourceFileColumn']
            student_file_column_name = mapping_row['StudentsFileColumn']
            target_column_name = mapping_row['TargetFileColumn']

            # Check if the source column exists in the sourceEventFileDF
            #print(source_column_name)
            source_column_value = row.get(source_column_name)
            if(source_column_value == 'Campus'):
                print(source_column_value + " " + getValueFromStudentRegistrarFile(row, student_file_column_name))
                
            # If source column value is 'Constant', set it directly

            if source_column_name == "Constant":
                #print("From Constant")
                target_row[target_column_name] = constants_dict[target_column_name]['Value']
            elif source_column_value is not None:
                #print("From source file")
                target_row[target_column_name] = source_column_value
            else:
                #print(source_column_name + '  student column name=' + student_file_column_name)
                # If the source column value is None or the column doesn't exist, fetch the value
                target_row[target_column_name] = getValueFromStudentRegistrarFile(row, student_file_column_name)
        #print(target_row)
        # Append target_row to target_file
        target_file[index] = target_row

    return pd.DataFrame.from_dict(target_file, orient='index')
```
